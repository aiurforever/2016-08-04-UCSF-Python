```python
# Find GDP per capita for all european countries after 1985
europe_df = df[df["continent"] == "Europe"]
needed_cols = ["country"]

for col in df.columns:
    if ("gdp" in col):
        split_list = col.split('_')
        if (int(split_list[1]) > 1985):
            needed_cols.append(col)

print(needed_cols)
print(europe_df[needed_cols])
```
